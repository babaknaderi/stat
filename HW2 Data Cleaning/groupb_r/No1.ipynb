```r
library(readxl)   # read in spreadsheet
library(dplyr)    # pipelining operations
library(reshape2) # used for clearing data points of NA values
```

```r
#load data
speech_quality_crowdsourcing <- read_excel("datasets/DB04_speech_quality_crowdsourcing_dataset.xlsx", sheet = 3)
speech_quality_crowdsourcing$cond = substr(speech_quality_crowdsourcing$files,6,8)
```

```r
# remove NA's and merge data
p <- select(speech_quality_crowdsourcing, starts_with("P"), cond)
crowd <- melt(p, id.vars=c('cond'))
crowd <- na.omit(crowd)
```

```r
crowd_all <- crowd %>% 
group_by(cond) %>% 
summarise(Mean = mean(value, na.rm = TRUE), Std_Dev = sd(value, na.rm = TRUE), n = n()) %>% 
mutate(CI95=Std_Dev/sqrt(n) * qt(0.95, df=n-1)) %>%
mutate(Lower_CI_Bound = Mean - qt(1 - (0.05 / 2), n - 1) * (Std_Dev / sqrt(n)),
        Upper_CI_Bound = Mean + qt(1 - (0.05 / 2), n - 1) * (Std_Dev / sqrt(n))) %>% # confidence interval
arrange(cond)                


crowd_all 
```
