=>    (new - 1) / (5 - 1) = (old - 1) / (7 - 1) <br>
<=>  6 * (new - 1)       = 4 * (old - 1) <br>
<=>  6 * new - 6         = 4 * old - 4 <br>
<=>  6 * new             = 4 * old + 2 <br>
<=>  new                 = 2/3 * old + 1/3

```r
DB01 <- read_excel("datasets/DB01_gaming_video_quality_dataset.xlsx")
```

# Transform all VQ data of 7-point scale to 5-point scale

```r
ScaledDB01 <- DB01 %>%
    mutate(VQ = (2/3) * VQ + 1/3)

ScaledDB01
```

# Normalization to [1, 4.5]

```r
minimumVQ = (ScaledDB01 %>% filter(VQ == min(VQ)))$VQ[1]
maximumVQ = (ScaledDB01 %>% filter(VQ == max(VQ)))$VQ[1]

paste0('minimum before normalization: ', minimumVQ)
paste0('maximum before normalization: ', maximumVQ)

NormalizedDB01 <- ScaledDB01 %>% mutate(VQ = ((VQ - minimumVQ) / (maximumVQ - minimumVQ)) * 3.5 + 1)

minimumVQ = (NormalizedDB01 %>% filter(VQ == min(VQ)))$VQ[1]
maximumVQ = (NormalizedDB01 %>% filter(VQ == max(VQ)))$VQ[1]

paste0('minimum before normalization: ', minimumVQ)
paste0('maximum before normalization: ', maximumVQ)

NormalizedDB01
```
