# Task 2: Calculate a new parameter called bitperpixel
## Transform resolution to number of pixel: e.g. 1080p => 1920x1080 = 2073600. You can reduce the number of parameters by making a new parameter ”bit per pixel” = (bitrate)/(number of pixel)

```r
# install.packages('dplyr')      # processing 
# install.packages('gdata')      # file reading
```

```r
library(dplyr)     # processing
library(readxl)    # reading in data
```

```r
# read in data set
gaming_data <- read_excel("datasets/DB01_gaming_video_quality_dataset.xlsx")
```

```r
# feature transformation: compute bits per pixel for every record
gaming_data_feature_transformation <- gaming_data %>% 
                                        mutate(Number_Of_Pixels = ifelse(Resolution == 480, 480*720, 
                                                                    ifelse(Resolution == 720, 720*1280, 
                                                                          ifelse(Resolution == 1080, 1080*1920, -1)
                                                                          )
                                                                  )
                                              ) %>%
                                        mutate(Bits_Per_Pixel = Bitrate / Number_Of_Pixels) %>%
                                        select(Game, Condition, Number_Of_Pixels, Bits_Per_Pixel)

# throw away unnecessary duplicates
unique(gaming_data_feature_transformation) %>% arrange(Game, Condition)
```
