# Task 1: Calculate MOS values
## Calculate as learned before the MOS values of VQ, VF, VU, and VD for each conditions separated for each game

```r
# install.packages('dplyr')      # processing 
# install.packages('gdata')      # file reading
```

```r
library(dplyr)     # processing
library(readxl)    # reading in data
```

```r
# read in data set
gaming_data <- read_excel("datasets/DB01_gaming_video_quality_dataset.xlsx")
```

```r
# compute MOS values for VQ, VF, VU, VD
mos_per_condition_and_game <- gaming_data %>%
                              group_by(Game, Condition) %>%
                              summarise(meanVQ = mean(VQ), meanVF = mean(VF), meanVU = mean(VU), meanVD = mean(VD)) %>%
                              select(Game, Condition, meanVQ, meanVF, meanVU, meanVD)

mos_per_condition_and_game
```
